```python
batch_size = 1 
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator()

train_generator = train_datagen.flow_from_directory(
    su.train_path,
    target_size=(su.height, su.width),
    batch_size=batch_size,
    classes=su.categories,
    class_mode="categorical",
)
```

```python
from keras.applications import vgg19
from keras import backend as K

for data_batch, labels_batch in train_generator:
    data_batch = data_batch
    print(data_batch.shape)
    break
```

```python
input_tensor = K.constant(data_batch)
model = vgg19.VGG19(
    input_tensor=input_tensor,
    weights='imagenet',
    include_top=False
)
```

```python
# recap the layers of VGG19
model.summary()
```

```python
# take the layers that define style
# data_batch.shape
outputs_dict = dict([(layer.name, layer.output) for layer in model.layers])

style_layers = ['block1_conv1',
                'block2_conv1',
                'block3_conv1',
                'block4_conv1',
                'block5_conv1']
```

```python
#outputs_dict
sess = tf.Session()
init = tf.global_variables_initializer()

with sess.as_default():
    init.run()
    for layer_name in style_layers:
        layer_features = outputs_dict[layer_name]
        #print (layer_features)
        style_reference_features = layer_features[0, :, :, :]
        #print (type (style_reference_features))
        print (style_reference_features.eval().shape)
```
